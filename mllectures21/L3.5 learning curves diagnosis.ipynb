# Learning Curves

To diagnose whether a model converges nicely without underfitting or overfitting we plot learning curves. On the y-axis we usually plot the loss and on the y-axis some hyperparameter.

# Hyperparameters

Hyperparameters are parameters that we set prior to learning, as opposed to other parameters that are learned during learning such as the coefficients for Linear Regression. Examples of hyper parameters are (when applicable):
- learning rate
- the number of epochs (passes over the data set) used for training
- the order of polynomials
- the maximum depth of a decision tree
- k for k-Nearest Neighbors
- the number of features when using a fixed strategy to add features
- regularization parameters (regularizations are ways to prevent overfitting)
- amount of data to use

```python
from ml import *
```

```python
data = dam(degree=2, scale=True)
model = linear_regression_sgd(data)
```

# Learning rate too small

This way learning takes a long time. Try increasing the learning rate. If that does not work, train for more epochs, since the model has not yet converged.

```python
model = linear_regression_sgd(data)
model.train(46, lr=0.03, report_frequency=1, bar=False)
model.history.plot(start=1)
```

# Learning rate to big

The learning curve should go down more smoothly. In fact for Linear Regression it usually should not bounce up, unless maybe very slightly when it almost converged. So lowering the learning rate should help.

```python
model = linear_regression_sgd(data)
model.train(500, lr=0.1, report_frequency=10, bar=False, verbose=False)
model.history.plot_train(start=3)
```

# Just right

It seems that this is a good setting for the learning rate. The model converges smoothly after about 200 epochs, after that the model does not really improve anymore.

```python
model = linear_regression_sgd(data)
model.train(500, lr=0.01, report_frequency=10, bar=False, verbose=False)
model.history.plot(start=3)
```
