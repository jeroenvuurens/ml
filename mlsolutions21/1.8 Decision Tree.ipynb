# Decision Tree

Decision trees iteratively split the population according to decision rules, e.g. gender, income, to identify groups that ideally all belong to the same class, e.g. political preference.

# Data 

For this, we can load the dataset with the `wines_binary` function in which we set the minimum quality for a wine to pass. A 2d scatter plot uses the colors green for the wines with a quality >= threshold and the color red for the wines that do not. We can roughly see that wines are better appreciated when containing more alcohol and being slightly more acidic.

```python
from ml import *
```

```python
data = wines_binary('quality', 'alcohol', 'pH', threshold=6)
```

```python
data.plot2d()
```

# Model, Training and Validation

The difference in code is just choosing a different model. You can use the decision_tree model and can set the max_depth parameter to allow the model to use more decision rules and thus create a more complex model. 

```python
model = decision_tree(data, max_depth=4)
model.train()
model.data.plot2d()
model.plot_boundary()
```
