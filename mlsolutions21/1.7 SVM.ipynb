# Support Vector Machines

SVMs are based on the idea of finding a hyperplane that best divides a dataset into two classes. The 'optimal' hyperplane is found by identifying the Support Vectors, i.e. data points that are closest to this hyperplane, and by finding the hyperlane that maximizes to distance (or margin) to every one of these Support Vectors.

# Data 

For this, we can load the dataset with the `wines_binary` function in which we set the minimum quality for a wine to pass. A 2d scatter plot uses the colors green for the wines with a quality >= threshold and the color red for the wines that do not. We can roughly see that wines are better appreciated when containing more alcohol and being slightly more acidic.

```python
from ml import *
```

```python
data = wines_binary('quality', 'alcohol', 'pH', threshold=6)
```

```python
data.plot2d()
```

# Model, Training and Validation

The difference in code is just choosing a different model. For a threshold of 6, we see that the SVM correctly classifies 70% of the wines based on their alcohol percentage and pH-value. 

```python
model = svm(data)
model.train()
model.data.plot2d()
model.plot_boundary()
```
