# Boston housing dataset

In the Boston Housing dataset, the median price of houses in neighborhoods of the city of Boston is given, with features like LSTAT - the percentage of 'lower class' people living in that neighborhood.

NB If you omit the features (e.g. lstat, nox) all the features are loaded. You can view the features with 

```python
from ml import *
```

```python
boston_pd().head()
```

```python
data = boston('lstat', 'nox', valid_perc=0.9)
data.plot()
```

#### What is your diagnosis of the plot below? If anything is wrong, how can you fix it.

```python
data = boston('lstat', 'nox', valid_perc=0.9)
model = linear_regression_sgd(data)
model.train(500, lr=0.000001, report_frequency=10, bar=False, verbose=False)
model.history.plot(start=1)
```

```python
data = boston('lstat', 'nox', valid_perc=0.2, degree=2, scale=True)
model = linear_regression_sgd(data)
model.train(4000, lr=0.003, report_frequency=10, bar=False, verbose=False)
model.history.plot(start=10)
```
