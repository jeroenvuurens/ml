#### Write an evaluation function to compute the accuracy and compute the accuracy for the model

```python
def evaluate(model, X, y):
    return sum( model.predict(X) == y ) / len(X)
```

```python
evaluate(model, data.train_X, data.train_y)
```

```python
evaluate(model, data.valid_X, data.valid_y)
```

#### Write the code to fit a decision tree to the titanic dataset. Make sure that you reload the dataset (because logistic_regression_sgd adds a bias and transposes the target variable). Evaluate this model.

```python
data = titanic()
model = DecisionTreeClassifier(max_depth=6)
model.fit(data.train_X, data.train_y)
evaluate(model, data.train_X, data.train_y)
```

```python
evaluate(model, data.valid_X, data.valid_y)
```

#### Write the code to fit an SVM to the titanic dataset. Make sure that you reload the dataset (because logistic_regression_sgd adds a bias and transposes the target variable). Evaluate this model.

```python
data = titanic(scale=True)
model = SVC(kernel='linear', C=0.025)
model.fit(data.train_X, data.train_y)
evaluate(model, data.train_X, data.train_y)
```
