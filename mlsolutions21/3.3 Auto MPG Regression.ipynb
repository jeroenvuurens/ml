# Autompg dataset

In the auto mpg dataset are cars. The target variable is Miles Per Gallon (mpg) which decsribes how efficient the car is. We can use several features to predict the mpg. In autompg we do not use origin and car name as features.

```python
data = autompg('horsepower')
data.plot()
```

```python
data = autompg('horsepower')
model = linear_regression_sgd(data)
model.train(500, lr=1e-4, report_frequency=10, bar=False, verbose=False)
model.history.plot(start=10)
```

```python
data = autompg('horsepower', degree=2, scale=True)
model = linear_regression_sgd(data)
model.train(2000, lr=1e-2, report_frequency=10, bar=False, verbose=False)
model.history.plot(start=10)
```
